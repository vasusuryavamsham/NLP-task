# Analysis

```python
# All imports

import string

import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.tokenize import sent_tokenize
import syllables
import re
```

```python
l_text = txt.lower()
# l_text
```

```python
string.punctuation
```

```python
# Text with no punctuations
cleared_text = l_text.translate(str.maketrans("","",string.punctuation))

# print(cleared_text)
```

```python
# Sentence Tokenizing
sents = sent_tokenize(txt)

t_sents = (len(sents))

# for i in sents:
#     print(i)

avg_sents = len(txt) / t_sents

t_sents,avg_sents
```

```python
# Word Tokenizing
tk_words = word_tokenize(cleared_text)

t_words = len(tk_words)
t_words
# print(tk_words)
```

```python
# Stop word loaded from nltk.corpus module
stp_words = set(stopwords.words("english"))

cleaned_words = [word for word in tk_words if word not in stp_words ]

# print(cleaned_words)
```

```python
total_words = len(cleaned_words)
total_words
```

```python
# loading all stop words
fl = open(r"C:\Users\vasu5\Desktop\assignment\stopwords\all_words.txt",'r')
stopwords = fl.readlines()
fl.close()
stop_words = []
for i in stopwords:
    
   
    clear = i.replace("\n","").replace(" ","")
    if("|" in i):
        clear = clear.split("|")
        stop_words.append(clear[0])
        stop_words.append(clear[1])
    else:
        stop_words.append(clear)

# stop_words
```

```python
# After removing all given stop words
final_words = [word for word in tk_words if word not in stop_words ]

# print(final_words)
len(final_words)
```

```python
fl = open(r"C:\Users\vasu5\Desktop\assignment\p-n\positive-words.txt",'r')
dt = fl.readlines()
pos = []
for i in dt:
    r = i.replace("\n","")
    pos.append(r)
    
# print(pos)
```

```python
fl = open(r"C:\Users\vasu5\Desktop\assignment\p-n\negative-words.txt",'r')
dt = fl.readlines()
neg = []
for i in dt:
    r = i.replace("\n","")
    neg.append(r)
    

# print(neg)
```

```python
# 1. positive score
pos_list = [] 
for i in pos:
    if i in final_words:
        pos_list.append(i)
pos_score = len(pos_list)
print(pos_score)
```

```python
# 2. negative score
neg_list = []
for i in neg:
    if i in final_words:
        neg_list.append(i)
neg_score = len(neg_list)
print(neg_score)
```

```python
# 3.  polarity score
polarity_score = (pos_score - neg_score)/ ((pos_score + neg_score) + 0.000001)
print(round(polarity_score,4))
```

```python
# 4. Subjectivity score
subjectivity_score = (pos_score + neg_score)/ ((total_words) + 0.000001)

print(round(subjectivity_score,4))
```

```python
# 5. AVG SENTENCE LENGTH

# Average Sentence Length = the number of words / the number of sentences

avg_sent_len = t_words/t_sents

print(round(avg_sent_len,2))
```

```python
# 8. AVG NUMBER OF WORDS PER SENTENCE

# Average Number of Words Per Sentence = the total number of words / the total number of sentences
avg_words_per_sent = t_words/t_sents

print(round(avg_words_per_sent,2))
```

```python
# 10. WORD COUNT

print(t_words)
```

```python
# 11. SYLLABLE PER WORD
syllables_count = []
# print(len(tk_words))
for word in tk_words:
    count = syllables.estimate(word)
    if word.endswith("es") or word.endswith("ed"):
        count -= 1
    if(count == 0):
        count = 1
    syllables_count.append(count)

# print(len(syllables_count))

# syllables per word = sum of count of all syllables / total number of words
sum_syllables = 0
for i in syllables_count:
    sum_syllables += i
syllables_per_word = sum_syllables / t_words
print(round(syllables_per_word,2))
```

```python
# 9. COMPLEX WORD COUNT
complex_words_count = 0
for i in syllables_count:
    if(i>3):
        complex_words_count += 1

print(complex_words_count)
        
```

```python
# 6. PERCENTAGE OF COMPLEX WORDS
# Percentage of Complex words = the number of complex words / the number of words 

percent_complex_words = complex_words_count / t_words
print(round(percent_complex_words,2))
```

```python
# 7. FOG INDEX
# Fog Index = 0.4 * (Average Sentence Length + Percentage of Complex words)
fog_index = 0.4 * (avg_sents + percent_complex_words)
print(round(fog_index,2))
```

```python
# 12. PERSONAL PRONOUNS
# To calculate Personal Pronouns mentioned in the text, we use regex to find the counts of the words - “I,” “we,” “my,” “ours,” and “us”. Special care is taken so that the country name US is not included in the list.

pattern = r'\b(I|we|my|ours|us)\b'

regex = re.compile(pattern, re.IGNORECASE)

matches = regex.findall(txt)

pp_count = 0

for match in matches:
    if(match != "US"):
        pp_count += 1

print(matches)
print(pp_count)
```

```python
# 13. AVG WORD LENGTH

# Sum of the total number of characters in each word/Total number of words




# or
count = 0
for i in tk_words:
    l = len(i)
    count += l
avg_word_len = count/t_words
print(round(avg_word_len,2))
```
